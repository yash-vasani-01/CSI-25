```python
sns.countplot(data=data,x='Sex')
plt.title('Distribution - Gender')
plt.show()
```

```python
sns.histplot(data = data ,x='Age' , bins=20, kde=True)
plt.title('Distribution - Age')
plt.show()
```

```python
sns.histplot(data = data ,x='Height' , bins=20, kde=True)
plt.title('Distribution - Height')
plt.show()
```

```python
sns.histplot(data = data ,x='Weight' , bins=20, kde=True)
plt.title('Distribution - Weight')
plt.show()
```

```python
sns.countplot(data=data,x='Medal')
plt.title('Medals Distribution')
plt.show()
```

```python
plt.figure(figsize=(25,8))
sns.countplot(data=data,x='Sport',hue='Medal')
plt.title('Medals Distribution by Sport')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(25,8))
sns.countplot(data=data,x='Year',hue='Medal')
plt.title('Medals Distribution by Sport')
plt.xticks(rotation=45)
plt.show()
```

```python
year_avg_age=data.groupby('Year')['Age'].mean().head(5)
year_avg_age.plot(kind='bar', figsize=(10, 5))
```

```python
sport_median_height = round(data.groupby('Sport')['Height'].median())
sport_median_height.head(10).plot(kind='bar', figsize=(10, 5))
```

```python
data.head(5)
```

```python
country_gender_count = data.groupby(['Team','Sex'])['Name'].count()
country_gender_count
```

```python
most_successful_athletes = data.groupby(['Name','Team'])['Medal'].count().sort_values(ascending=False).head(10)
most_successful_athletes.plot(kind='bar', figsize=(10, 5))
```

```python
country_gender_weight = data.groupby(['NOC', 'Sex'])['Weight'].mean()
country_gender_weight['IND'].plot(kind='bar', figsize=(10, 5))
```

```python
sport_event_count = data.groupby('Sport')['Event'].nunique().sort_values(ascending=False)
plt.figure(figsize=(20, 6))
sport_event_count.plot(kind='bar')
plt.title('Number of Uniqe Event per Sport')
plt.xticks(rotation = 90)
plt.xlabel('Sports')
plt.ylabel('Number of Uninq Events')
plt.show()
```

```python
year_participants_count = data.groupby('Year')['ID'].nunique().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
year_participants_count.plot(kind='bar')
plt.title('Number of Participants of year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

```python
country_ave_age = data.groupby('NOC')['Age'].mean().sort_values(ascending=False)
country_ave_age.head(10).plot(kind='bar')
plt.title('Top 10 Country With Highest Average Age of Participants')
plt.xlabel('NOC')
plt.ylabel('Average Age')
plt.xticks(rotation= 45)
plt.show()
```

```python
sns.boxplot(data=data ,  x='Season', y='Age')
plt.title('Distribution of Age by Season')
plt.xlabel('Season')
plt.ylabel('Age')
plt.show()
```

```python
sns.violinplot(data=data, x='Medal', y='Height', palette='Set2', hue='Sex')
plt.title('Distributions of Hights by Medals')
plt.xlabel('Medal')
plt.ylabel('Height')
plt.show()
```

```python
most_medals_country = data['NOC'].value_counts().idxmax()
most_medals_country
```

```python
tallest_athlete = data[data['Height'] == data['Height'].max()]
tallest_athlete[['ID', 'Name', 'Height', 'Sport']]
```

```python
haviest_athlete = data[data['Weight'] == data['Weight'].max()]
haviest_athlete[['ID', 'Name', 'Weight','Height', 'Sport']]
```

```python
sns.scatterplot(data=data, x='Height', y='Weight', hue='Medal', palette='Set1')
plt.title('Athlete Height vs Weight by Medal Status')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.legend(title='Medal')
plt.show()
```

```python
sns.scatterplot(data=data, x='Age', y='Weight', hue='Medal', palette='Set1')
plt.title('Athlete Age vs Weight by Medal Status')
plt.xlabel('Age')
plt.ylabel('Height')
plt.legend(title='Medal')
plt.show()
```
